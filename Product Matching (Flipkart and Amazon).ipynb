Polyfuzz is used for string Matching

That how much 2 string matches each other

```python
#pip install polyfuzz
import polyfuzz as pf
```

```python
# Matching a column through TF-Idf
model = pf.PolyFuzz("TF-IDF").match(data['product_name_in_amazon'].values.tolist(), data['product_name_in_flipkart'].values.tolist())
```

```python
similar = pd.DataFrame(model.get_matches())
```
